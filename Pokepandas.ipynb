```python
import numpy as np
def split_train_test(data, test_ratio):
    shuffled_indicies = np.random.permutation(len(data))
    test_set_size =int(len(data)*test_ratio)
    test_indicies = shuffled_indicies[:test_set_size]
    train_indicies = shuffled_indicies[test_set_size:]
    return data. iloc[train_indicies], data.iloc[test_indicies]
```

```python
train_set, test_set = split_train_test(pdata, 0.2) #20% is the test_set
print(len(pdata))
print(len(train_set))
print(len(test_set))
```

```python
#We can basically do the above with sklearn with seed feature
from sklearn.model_selection import train_test_split
set_train, set_test = train_test_split(pdata, test_size =0.2, random_state =42)
print(len(set_train))
print(len(set_test))
```

```python
# Organize sum from deceding order
pdata.sort_values(["sum"], axis=0, ascending=False, inplace=True) #sorts two cols into decending
pdata['sum'].head(20)
```

```python
#???  Grabbing the highest sums only 
X = pdata['Generation'].head(80)
y = pdata['sum'].head(80)
# pdata.iloc[1:5] ?

X_b = np.c_[np.ones((80, 1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
X_new = np.array([[0], [6]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
import matplotlib.pyplot as plt

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 7, 500, 900])
plt.show()  #Displays top 80 pokemon sums
```
